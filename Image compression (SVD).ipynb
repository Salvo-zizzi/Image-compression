```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import svd
```

```python
# Load an image (grayscale)
image = plt.imread('/Users/salvatorezizzi/Desktop/IMG_20240914_150157810_HDR.jpg')  # Replace with your image path
image = image[:, :, 0]  # Convert to grayscale if it's RGB
```

```python
# Perform SVD
U, S, Vt = svd(image, full_matrices=False)
```

```python
# Choose the number of singular values to keep
k = 40  # You can adjust this value
U_k = U[:, :k]
S_k = np.diag(S[:k])
Vt_k = Vt[:k, :]
```

```python
# Reconstruct the image
compressed_image = np.dot(U_k, np.dot(S_k, Vt_k))
```

```python
# Display original and compressed images
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Compressed Image (k={})'.format(k))
plt.imshow(compressed_image, cmap='gray')
plt.axis('off')

plt.show()
```
